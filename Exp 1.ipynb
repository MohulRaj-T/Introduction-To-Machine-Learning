```python
def prime(num):
  num=int(input("Enter a number:"))
  if num<=1:
    return False
  for i in range(2,int(num**0.5)+1):
    if num%i==0:
      print("Is a prime number")
      return
    print("Is not a prime number")
prime(num)
```
